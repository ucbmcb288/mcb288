```python
# load dataframe with expression data

datafile = "data/barton/Barton_combined.txt"
ye = pd.read_csv(datafile)   # the sep='\t' tells pandas that it is a tab separated file

ye = ye.set_index('Gene')
```

```python
# this comment gets all of the column names and puts them into a list

exps = {}

exps['wt'] = [e for e in ye.columns if e.startswith('WT')] 

exps['mut'] = [e for e in ye.columns if e.startswith('Snf2')] 

exps['all'] = exps['wt'] + exps['mut']
```

```python
# normalize all experimental columns

ye['all_gmean']  = gmean(ye[exps['all']].replace(0,np.nan).values,axis=1)

df = ye[ye['all_gmean'] > 10]

for c in exps['all']:
    norm = np.median(df[c]/df['all_gmean'])
    ye[c] = ye[c] / norm
```

```python
# because the normalization affects the means I often go through a few cycles 
```

```python
for cycles in range(0,10):
    ye['all_gmean']  = gmean(ye[exps['all']].replace(0,np.nan).values,axis=1)
    
    df = ye[ye['all_gmean'] > 10]

    for c in exps['all']:
        norm = np.median(df[c]/df['all_gmean'])
        ye[c] = ye[c] / norm
```

```python
# computer row stats

for d in ['all','wt','mut']:
    ye[d + '_mean']   = ye[exps[d]].mean(axis=1)
    ye[d + '_median'] = ye[exps[d]].median(axis=1)
    ye[d + '_var']    = ye[exps[d]].var(axis=1)
    ye[d + '_std']    = ye[exps[d]].std(axis=1)
    ye[d + '_gmean']  = gmean(ye[exps[d]].replace(0,np.nan).values,axis=1)
```

```python
ye['med_ratio'] =  ye['mut_median'] / ye['wt_median'] 
ye['true_de'] = np.log(ye['med_ratio'])/np.log(2)
ye['abs_true_de'] = np.abs(ye['true_de'])
```

```python
df = ye.copy()
df = df[df[exps['all']].min(axis=1) > 10]
df = df.sort_values('wt_mean',ascending=False)

for c in exps['all']:
    df[c] = df[c] - df['wt_mean']
    
mut_ind = []
mut = []
wt_ind = [] 
wt = []

row = 0
for i,r in df.head(200).iterrows():
    for c in exps['mut']:
        mut_ind.append(row+1)
        mut.append(r[c])
    for c in exps['wt']:
        wt_ind.append(row+1)
        wt.append(r[c])
    row = row + 1
        
fig = plt.figure(figsize = (10,.1*max(wt_ind)))
ax = fig.add_subplot(1,1,1)
ax.set_ylim(max(wt_ind + mut_ind),0)
#ax.set_xscale("log", nonposx='clip')
ax.set_xlabel("expression level")
ax.set_ylabel("gene - sorted by expression")
ax.scatter(wt,wt_ind,s=12,alpha=.25,c='r',edgecolor='none')
#ax.scatter(mut,mut_ind,s=12,alpha=.25,c='b',edgecolor='none')
```

```python
def nbinom_muvar(mu,var):
    if var > mu:
        p = (var - mu) / var
        r = mu**2 / (var - mu)
        #print (mu,var,r,1-p)
        return (r,1-p)
    else:
        print("variance must be greater than mean")
```

```python
n0 = 1000000
p0 = .001

nb0 = nbinom(n = n0,p = 1-p0)
a0 = nb0.mean()
v0 = nb0.var()
x = np.arange(0,2*a0,int(a0/100))

plt.scatter(x,nb0.pmf(x))

cp = nbinom_muvar(1000,100000)
print (cp)
nb1 = nbinom(n=cp[0],p=cp[1])
plt.scatter(x,nb1.pmf(x))

p0 = nb0.pmf(x)
p0 = p0 / np.sum(p0)
a0 = np.average(x,weights=p0)
v0 = np.average((x-a0)**2,weights=p0)
p1 = nb1.pmf(x)
p1 = p1 / np.sum(p1)
a1 = np.average(x,weights=p1)
v1 = np.average((x-a1)**2,weights=p1)

print (a0,v0,a1,v1)
```
