```python
import scipy
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
hl = 2.5
r = np.log(0.5)/hl
d = 1000
doses = 100
i = 6
decay = np.exp(r*i)

x = np.linspace(0,doses*i,100*doses)
yt = np.zeros(len(x))

for dose in range(0,doses):
    xd = dose * i
    yd = np.where(x >= xd, d * np.exp(r * (x-xd)),0)
    yt += yd
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,yt,label="6 hour interval")
plt.plot(x,yt3,label="3 hour interval")
plt.xlim(0,200)
plt.xlabel("Time (hours)")
plt.ylabel("Acetominophen in blood (mg)")
#plt.title("Dose: %d mg, Half-life: %.2f hours, Interval between doses: %d hours" % (d,hl,i))
plt.legend()
```
