# Hierarchical clustering

## Types of Hierarchical clustering
### (1) agglomerative: start from tips of the tree and group branches, recursively progressing down to the trunk
### (2) divisive: start from the trunk, and split off subgroups, recursively applying the same method to each new branch.


```python
# import standard libraries
import pandas as pd                                    # data frames  
import numpy as np                                     # matrices 
import matplotlib.pyplot as plt                        # plotting 
import seaborn as sns; sns.set()                       # better plotting
from sklearn.model_selection import train_test_split   # training-vs-testing
from sklearn.linear_model import LinearRegression      # the linear regression model
from sklearn import metrics
%matplotlib inline
```

```python
# create dataset
from sklearn.datasets import make_blobs 
X, y = make_blobs(n_samples = 150, 
                  n_features = 2,
                  centers = 4,
                  cluster_std = 2, 
                  shuffle=True, 
                  random_state=0)
```

```python
# visualization: color each group differently
marker_list = ["s","o","v","d","s","o","v","d"]
color_list = ['red','orange','yellow','green',"blue","indigo","violet", "lightgreen","lightblue"]
for i in range(4):
    plt.scatter( X[y ==i, 0], X[y ==i, 1], 
                s=50, c=color_list[i], marker = marker_list[i], edgecolor='black',label = i)
# add the centroids
#plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1],
#           s = 250, marker = "*", c='red', edgecolor='black', label='centroids')
# add legend
plt.legend()
plt.show()
```

```python
# agglomerative clustering
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as shc
```

```python
# make a dendrogram.  
# "ward" -- group progressively to have minimum variance at each node.
plt.figure(figsize=(10,7))
plt.title("Dendrogram")
dend = shc.dendrogram(shc.linkage(X, method='ward'))
```

```python
# this also defines a clustering, given either 
#        a number of clusters
#        a threshold cutoff
model = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage="ward")
model.fit(X)
y = model.labels_
print(y)
```

```python
# ’single’ single link, connect the next closest pair
# "complete" join clusters from level n-1 that have the smallest maximal distance
# 'average' join clusters from level n-1 with the smallest average distance (aka UPGMA)
# "weighted"  
# 'centroid'
#  'median'
```
